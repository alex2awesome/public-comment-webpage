__Anthropic__

__Similar agencies with potential variation:__

* EPA vs. FDA - Both do highly technical rulemaking, but EPA deals with more politically polarized issues (climate, air quality) while FDA often has industry-focused comments on drug approvals. Does political salience change what counts as "substantive"?
* OSHA vs. MSHA (Mine Safety and Health Administration) - Both handle workplace safety, similar statutory mandates, but very different constituencies (general labor vs. mining industry). Do specialized industries get different treatment?
* Different financial regulators: SEC, CFPB, and FDIC all regulate aspects of finance but have different founding eras, political controversy levels, and commenting communities.

__Agencies with interesting structural differences:__

* NHTSA vs. FAA - Both transportation safety, but FAA has much more technical/industry dominance in comments while NHTSA gets more consumer advocacy.
* FCC - Gets enormous comment volumes (remember net neutrality?) that are often form letters. How do they distinguish "substantive" when flooded?
* HHS (excluding FDA) - Tackles both highly technical Medicare payment rules and hot-button issues like reproductive health or the ACA.

__Cross-administration comparison targets:__

* EPA is the obvious choice - swings dramatically between administrations
* Department of Education - Title IX rules, student loans
* Department of Labor - overtime rules, independent contractor classification

__Gemini__

* __SEC vs. Treasury (IRS)__
If you are looking for an agency that mirrors the SEC’s "economic substance" but with a different enforcement flavor, look at the Department of the Treasury, specifically the IRS. The Comparison: Both deal with the "Truth" of financial transactions. However, SEC comments are often about disclosure (what must the public know?), while IRS/Treasury comments are about definitions (how is this money classified?).

* __EPA vs. DHS__
The Department of Homeland Security (DHS) is the "EPA of the current administration" in terms of volume and intensity. The Comparison: EPA rules are often "Anti-Pollution" (restrictive on industry). DHS rules are currently "Border-focused" (restrictive on movement). Why it’s interesting: Both agencies deal with Externalities. EPA deals with chemical runoff; DHS deals with human migration. Both receive massive amounts of "form-letter" comments from NGOs. If you submit a "Moral/Humanitarian" argument to the EPA, it’s ignored. If you submit it to the DHS regarding a border rule, is it also ignored? Or does the DHS have a different "substantive" threshold for comments that cite "National Security" vs. "Civil Rights"?
