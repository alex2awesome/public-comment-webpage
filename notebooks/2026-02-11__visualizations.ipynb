```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Wedge, Circle
from collections import defaultdict
import math

# Helper function to get agency data
def get_agency_data(agency_code):
    """Get document and comment counts for an agency, with defaults for uncollected."""
    comment_counts, gov_counts = 0, 0
    # If agency has specific counts, use them
    if agency_code in gov_doc_counts:
        gov_counts = gov_doc_counts[agency_code]

    if agency_code in aggregated_comment_counts:
         comment_counts = aggregated_comment_counts[agency_code]
    
    # Otherwise, use defaults if collected, or 0 if not collected
    if (gov_counts == 0) and data['collected']:
        gov_counts = DEFAULT_GOV_DOCS
    if (comment_counts == 0) and data['collected']:
        comment_counts = DEFAULT_COMMENTS
    return gov_counts, comment_counts

# CONFIGURATION
HORIZONTAL_SPACING = 1.2
VERTICAL_SPACING = .5
VERTICAL_OFFSET = -1.5
COLS = 3

# ── GROUP BY TOPIC ────────────────────────────────────────────────────────────

topic_to_agencies = defaultdict(list)
for agency, data in agency_mapping.items():
    primary_topic = data["topic_cluster"].split(" / ")[0]
    topic_to_agencies[primary_topic].append(agency)

topic_stats = {}
for topic, agencies in topic_to_agencies.items():
    total_docs = sum(get_agency_data(a)[0] for a in agencies)
    total_comments = sum(get_agency_data(a)[1] for a in agencies)
    collected = sum(1 for a in agencies if agency_mapping[a]["collected"])
    collected_agencies = [a for a in agencies if agency_mapping[a]["collected"]]
    topic_stats[topic] = {
        'total_docs': total_docs,
        'total_comments': total_comments,
        'total_size': total_docs + total_comments,
        'collected': collected,
        'total': len(agencies),
        'collected_agencies': collected_agencies,
    }

sorted_topics = sorted(topic_stats.keys(),
                        key=lambda x: topic_stats[x]['total_size'], reverse=True)

# ── COLORS ────────────────────────────────────────────────────────────────────

topic_colors = {
    'Health & Medical': '#e74c3c',
    'Food Safety & Agriculture': '#27ae60',
    'Environmental & Natural Resources': '#16a085',
    'Emergency Management & Homeland Security': '#e67e22',
    'Transportation Safety': '#3498db',
    'Immigration & Border Security': '#9b59b6',
    'Law Enforcement & Justice': '#c0392b',
    'Worker Safety': '#f39c12',
    'Communications & Technology': '#1abc9c',
    'Research & Standards': '#95a5a6',
    'Financial Regulation': '#34495e',
}

# ── LAYOUT COMPUTATION ────────────────────────────────────────────────────────
# Goal: place topic bubbles in a grid with *dynamic* row heights so that rows
# with smaller bubbles don't waste vertical space.

cols = COLS
n_topics = len(sorted_topics)
rows = math.ceil(n_topics / cols)

# Outer-bubble radius is proportional to sqrt(total_size).
max_bubble = max(s['total_size'] for s in topic_stats.values())
MAX_RADIUS = 0.115          # maximum bubble radius in axis-fraction coords
MIN_RADIUS = 0.025          # minimum visible radius

def topic_radius(topic):
    size = topic_stats[topic]['total_size']
    if max_bubble == 0:
        return MIN_RADIUS
    r = MAX_RADIUS * math.sqrt(size / max_bubble)
    return max(r, MIN_RADIUS)

# Compute the maximum radius needed in each row so we can size rows properly.
row_max_radius = [0.0] * rows
for idx, topic in enumerate(sorted_topics):
    r = idx // cols
    row_max_radius[r] = max(row_max_radius[r], topic_radius(topic))

# Vertical: allocate space per row proportional to its max radius, plus padding
TITLE_RESERVE = 0.05    # space consumed by the title + subtitle text
BOTTOM_MARGIN = 0.06    # room for legend
LABEL_PAD = 0.035       # space above each bubble for text labels
ROW_GAP = 0.02          # fixed gap between rows

# VERTICAL_OFFSET is the user-configurable gap between the title area and
# the first row of bubbles (set in CONFIGURATION at the top).
top_used = TITLE_RESERVE + VERTICAL_OFFSET
available_height = 1.0 - top_used - BOTTOM_MARGIN

# Each row needs: label_pad + radius (above center) + radius (below center)
row_heights = [2 * r + LABEL_PAD for r in row_max_radius]
total_content = sum(row_heights)
total_gaps = ROW_GAP * (rows - 1)

# If content + gaps < available, add modest inter-row padding (extra space
# falls below the last row and gets clipped by the dynamic y-axis bounds).
used = total_content + total_gaps
if used < available_height:
    extra = available_height - used
    extra_per_gap = extra / max(rows - 1, 1)
    extra_per_gap = min(extra_per_gap, 0.025)  # cap extra gap
    effective_gap = ROW_GAP + extra_per_gap
else:
    effective_gap = ROW_GAP

# Content always starts right at the top of the available area so that
# VERTICAL_OFFSET directly controls the gap between title and first row.
start_y = 1.0 - top_used

# Compute each row's center-y, working top to bottom
row_cy = [0.0] * rows
cursor_y = start_y
for r_idx in range(rows):
    cursor_y -= (LABEL_PAD + row_max_radius[r_idx])  # top of row to center
    row_cy[r_idx] = cursor_y
    cursor_y -= row_max_radius[r_idx]  # center to bottom of row
    if r_idx < rows - 1:
        cursor_y -= effective_gap

# ── FIGURE ────────────────────────────────────────────────────────────────────

fig, ax = plt.subplots(figsize=(22, 14))
fig.patch.set_facecolor('#fafafa')
ax.set_facecolor('#fafafa')

fig.suptitle(
    'Federal Agency Data Collection by Topic Cluster',
    fontsize=22, fontweight='bold', y=0.97,
)
ax.text(0.5, 0.97,
        'Circle area ∝ total records  ·  Left half (dark) = gov documents  ·  Right half (light) = public comments',
        transform=ax.transAxes, ha='center', va='top', fontsize=12, color='#555555')

# Global max for agency-level circle scaling
all_agency_totals = []
for topic in sorted_topics:
    for a in topic_stats[topic]['collected_agencies']:
        d, c = get_agency_data(a)
        all_agency_totals.append(d + c)
global_max_agency = max(all_agency_totals) if all_agency_totals else 1

AGENCY_MAX_R = 0.032   # max radius of an agency sub-circle
AGENCY_MIN_R = 0.008   # min visible

def agency_radius(total):
    if global_max_agency == 0 or total == 0:
        return AGENCY_MIN_R
    r = AGENCY_MAX_R * math.sqrt(total / global_max_agency)
    return max(r, AGENCY_MIN_R)

# ── DRAW ──────────────────────────────────────────────────────────────────────

for idx, topic in enumerate(sorted_topics):
    r_idx = idx // cols
    c_idx = idx % cols

    cx = (c_idx + 0.5) / cols
    cy = row_cy[r_idx]
    radius = topic_radius(topic)

    stats = topic_stats[topic]
    color = topic_colors.get(topic, '#999999')

    # Outer translucent bubble
    outer = Circle((cx, cy), radius,
                   facecolor=color, alpha=0.10,
                   edgecolor=color, linewidth=2.5, linestyle='--', zorder=1)
    ax.add_patch(outer)

    # Topic label
    ax.text(cx, cy + radius + 0.012, topic,
            ha='center', va='bottom', fontsize=10, fontweight='bold', color=color)
    ax.text(cx, cy + radius + 0.003,
            f"{stats['collected']}/{stats['total']} agencies  ·  "
            f"{stats['total_docs']:,} docs  ·  {stats['total_comments']:,} comments",
            ha='center', va='top', fontsize=6.5, style='italic', color=color, alpha=0.8)

    # --- Agency sub-circles (only collected) ---
    collected = stats['collected_agencies']
    n = len(collected)
    if n == 0:
        ax.text(cx, cy, 'no data\ncollected', ha='center', va='center',
                fontsize=7, color='#aaaaaa', style='italic')
        continue

    # Sort agencies by total size descending (biggest in center / top)
    collected = sorted(collected, key=lambda a: sum(get_agency_data(a)), reverse=True)

    # Place agencies inside the outer bubble in a circle (or center if 1)
    orbit_radius = radius * 0.55 if n > 1 else 0.0
    angle_offset = math.pi / 2  # start at top

    for i, agency in enumerate(collected):
        docs, comments = get_agency_data(agency)
        total = docs + comments
        ar = agency_radius(total)

        if n == 1:
            ax_, ay_ = cx, cy
        else:
            angle = angle_offset + 2 * math.pi * i / n
            ax_ = cx + orbit_radius * math.cos(angle)
            ay_ = cy + orbit_radius * math.sin(angle)

        if total == 0:
            c0 = Circle((ax_, ay_), ar * 0.5, color='#cccccc', alpha=0.5, zorder=2)
            ax.add_patch(c0)
            ax.text(ax_, ay_, agency.upper(), ha='center', va='center',
                    fontsize=5.5, fontweight='bold', color='#888888', zorder=4)
            continue

        # --- Proportional split ---
        # docs_fraction of the full circle goes to documents (drawn as LEFT starting wedge)
        # comments get the rest.  We anchor "left = docs, right = comments" visually:
        #   docs wedge spans from  90° + half_comments_angle  to  90° + half_comments_angle + docs_angle
        # Simpler approach: docs sweep starts at top-left, comments sweep fills the rest.
        docs_frac = docs / total
        docs_angle = docs_frac * 360  # degrees

        # Draw docs wedge (darker colour) – sweep from 90° going counter-clockwise
        # Convention: 90° is top; we put docs on the LEFT side.
        # Wedge(center, r, theta1, theta2) draws CCW from theta1 to theta2.
        # Left half of a circle = 90° to 270° (going through 180°).
        # We want docs to fill *proportionally* from top going left.
        docs_start = 90
        docs_end = 90 + docs_angle  # goes counter-clockwise (left then down)

        docs_wedge = Wedge((ax_, ay_), ar, docs_start, docs_end,
                           facecolor=color, alpha=0.92,
                           edgecolor='white', linewidth=0.8, zorder=3)
        ax.add_patch(docs_wedge)

        # Comments wedge (lighter colour) – the rest
        rgb = mcolors.to_rgb(color)
        lighter = tuple(min(1.0, c * 1.35 + 0.15) for c in rgb)
        lighter_hex = mcolors.to_hex(lighter)

        comments_wedge = Wedge((ax_, ay_), ar, docs_end, docs_start + 360,
                               facecolor=lighter_hex, alpha=0.92,
                               edgecolor='white', linewidth=0.8, zorder=3)
        ax.add_patch(comments_wedge)

        # Thin outline for clarity
        outline = Circle((ax_, ay_), ar, fill=False,
                          edgecolor=color, linewidth=0.6, alpha=0.5, zorder=3.5)
        ax.add_patch(outline)

        # Label
        ax.text(ax_, ay_, agency.upper(),
                ha='center', va='center', fontsize=5.8, fontweight='bold',
                color='white', zorder=4,
                path_effects=[
                    __import__('matplotlib.patheffects', fromlist=['withStroke']).withStroke(linewidth=2, foreground=color)
                ])

# ── AXES ──────────────────────────────────────────────────────────────────────
# Compute tight vertical bounds from actual content
y_min = min(row_cy[r] - row_max_radius[r] for r in range(rows)) - 0.05
y_max = 1.02
ax.set_xlim(-0.02, 1.02)
ax.set_ylim(y_min, y_max)
ax.set_aspect('equal')
ax.axis('off')

# ── LEGEND ────────────────────────────────────────────────────────────────────
legend_elements = [
    mpatches.Patch(facecolor='gray', alpha=0.10, edgecolor='gray',
                   linewidth=2.5, linestyle='--',
                   label='Topic cluster (size ∝ total records)'),
    mpatches.Patch(color='gray', alpha=0.9,
                   label='Agency — gov documents (proportional)'),
    mpatches.Patch(color='lightgray', alpha=0.9,
                   label='Agency — public comments (proportional)'),
]
ax.legend(handles=legend_elements, loc='lower center', ncol=3,
          bbox_to_anchor=(0.5, -0.04), fontsize=11, frameon=True,
          fancybox=True, shadow=False, edgecolor='#cccccc')

plt.tight_layout(rect=[0, 0.02, 1, 0.95])
plt.savefig('agency_by_topic_nested.png', dpi=200, bbox_inches='tight',
            facecolor=fig.get_facecolor())
print("Saved → agency_by_topic_nested.png")

# ── SUMMARY TABLE ─────────────────────────────────────────────────────────────
print("\n=== DATA COLLECTION BY TOPIC CLUSTER ===")
for topic in sorted_topics:
    s = topic_stats[topic]
    pct = s['total_docs'] / s['total_size'] * 100 if s['total_size'] else 0
    print(f"\n{topic}:")
    print(f"  Collected: {s['collected']}/{s['total']} agencies")
    print(f"  Documents: {s['total_docs']:,}  ({pct:.1f}% of records)")
    print(f"  Comments:  {s['total_comments']:,}")
    print(f"  Agencies:  {', '.join(a.upper() for a in s['collected_agencies'])}")
```

```python
HORIZONTAL_SPACING = 1.2
VERTICAL_SPACING = 1.5
VERTICAL_OFFSET = 1.5


# Group agencies by parent
parent_to_children = defaultdict(list)
for agency, data in agency_mapping.items():
    parent = data["head_agency"]
    parent_to_children[parent].append(agency)

# Calculate parent agency stats
parent_stats = {}
for parent, children in parent_to_children.items():
    total_docs = 0
    total_comments = 0
    for c in children:
        docs, comments = get_agency_data(c)
        total_docs += docs
        total_comments += comments
    
    collected_count = sum(1 for c in children if agency_mapping[c]["collected"])
    
    parent_stats[parent] = {
        'total_docs': total_docs,
        'total_comments': total_comments,
        'total_size': total_docs + total_comments,
        'collected_count': collected_count,
        'total_count': len(children)
    }

# Sort parents by size for better layout
sorted_parents = sorted(parent_stats.keys(), key=lambda x: parent_stats[x]['total_size'], reverse=True)

# Create figure with subplots - one per parent agency
n_parents = len(sorted_parents)
cols = 3
rows = math.ceil(n_parents / cols)

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 5))
fig.suptitle('Federal Agency Hierarchy\nNode size = Documents + Comments collected', 
             fontsize=20, fontweight='bold', y=0.995)

if rows == 1:
    axes = axes.reshape(1, -1)

# Flatten axes for easier iteration
axes_flat = axes.flatten()

# Color scheme
color_collected = '#2ecc71'  # Green
color_not_collected = '#e74c3c'  # Red
color_parent = '#3498db'  # Blue
topic_cluster_colors = {
    'Health & Medical': '#e91e63',
    'Food Safety & Agriculture': '#8bc34a',
    'Environmental & Natural Resources': '#4caf50',
    'Emergency Management & Homeland Security': '#ff9800',
    'Transportation Safety': '#2196f3',
    'Immigration & Border Security': '#9c27b0',
    'Law Enforcement & Justice': '#f44336',
    'Worker Safety': '#ff5722',
    'Communications & Technology': '#00bcd4',
    'Research & Standards': '#9e9e9e',
    'Financial Regulation': '#607d8b',
}

# Draw each parent agency tree
for idx, parent in enumerate(sorted_parents):
    ax = axes_flat[idx]
    children = parent_to_children[parent]
    
    # Calculate layout positions
    # Parent at top center
    parent_x, parent_y = 0.5, 0.9
    
    # Children arranged in arc below
    n_children = len(children)
    if n_children == 1:
        child_positions = [(0.5, 0.3 - VERTICAL_OFFSET)]
    else:
        child_positions = []
        for i, child in enumerate(children):
            # Spread children in a semicircle - use HORIZONTAL_SPACING and VERTICAL_SPACING
            angle = math.pi * (i / (n_children - 1))  # From 0 to pi
            x = 0.5 + 0.35 * HORIZONTAL_SPACING * math.cos(angle)
            y = 0.3 - 0.25 * VERTICAL_SPACING * math.sin(angle) - VERTICAL_OFFSET
            child_positions.append((x, y))
    
    # Draw edges
    for i, child in enumerate(children):
        cx, cy = child_positions[i]
        ax.plot([parent_x, cx], [parent_y, cy], 'k-', alpha=0.3, linewidth=1, zorder=1)
    
    # Draw parent node (semi-circles for docs and comments)
    parent_docs = parent_stats[parent]['total_docs']
    parent_comments = parent_stats[parent]['total_comments']
    parent_size = math.sqrt(parent_stats[parent]['total_size']) * 3
    
    # Left semi-circle (documents) - blue shade
    parent_docs_wedge = Wedge((parent_x, parent_y), parent_size/1000, 90, 270, 
                              color='#2980b9', alpha=0.9, zorder=3, edgecolor='white', linewidth=2)
    ax.add_patch(parent_docs_wedge)
    
    # Right semi-circle (comments) - lighter blue shade
    parent_comments_wedge = Wedge((parent_x, parent_y), parent_size/1000, 270, 90,
                                  color='#3498db', alpha=0.9, zorder=3, edgecolor='white', linewidth=2)
    ax.add_patch(parent_comments_wedge)
    
    ax.text(parent_x, parent_y, parent, ha='center', va='center', 
            fontsize=14, fontweight='bold', color='white', zorder=4)
    
    # Add parent stats
    ax.text(parent_x, parent_y - parent_size/1000 - 0.05, 
            f"{parent_stats[parent]['collected_count']}/{parent_stats[parent]['total_count']} collected\n"
            f"{parent_stats[parent]['total_docs']:,} docs | {parent_stats[parent]['total_comments']:,} comments",
            ha='center', va='top', fontsize=8, style='italic')
    
    # Draw children nodes
    for i, child in enumerate(children):
        cx, cy = child_positions[i]
        data = agency_mapping[child]
        
        # Get actual data for this agency
        docs, comments = get_agency_data(child)
        total = docs + comments
        
        if data['collected']:
            size = math.sqrt(total) * 1.5 if total > 0 else 30
            alpha = 0.8
        else:
            size = 30
            alpha = 0.5
        
        # Get topic cluster color (use first topic if multiple)
        topic = data['topic_cluster'].split(' / ')[0]
        edge_color = topic_cluster_colors.get(topic, '#999999')
        
        if data['collected']:
            # Left semi-circle (documents) - darker green
            docs_wedge = Wedge((cx, cy), size/1000, 90, 270,
                              color='#27ae60', alpha=alpha, edgecolor=edge_color, linewidth=2, zorder=2)
            ax.add_patch(docs_wedge)
            
            # Right semi-circle (comments) - lighter green
            comments_wedge = Wedge((cx, cy), size/1000, 270, 90,
                                  color='#2ecc71', alpha=alpha, edgecolor=edge_color, linewidth=2, zorder=2)
            ax.add_patch(comments_wedge)
        else:
            # Uncollected - single color circle
            circle = plt.Circle((cx, cy), size/1000, color=color_not_collected, alpha=alpha, 
                               edgecolor=edge_color, linewidth=2, zorder=2)
            ax.add_patch(circle)
        
        # Label
        ax.text(cx, cy, child.upper(), ha='center', va='center', 
                fontsize=7, fontweight='bold', color='white', zorder=4)
        
        # Topic cluster below (truncated)
        topic_short = data['topic_cluster'][:20] + '...' if len(data['topic_cluster']) > 20 else data['topic_cluster']
        ax.text(cx, cy - size/1000 - 0.03, topic_short, 
                ha='center', va='top', fontsize=6, style='italic')
    
    # Configure axes
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    ax.set_aspect('equal')

# Hide unused subplots
for idx in range(n_parents, len(axes_flat)):
    axes_flat[idx].axis('off')

# Add legend
legend_elements = [
    mpatches.Patch(color='#2980b9', label='Parent Agency (left=docs, right=comments)'),
    mpatches.Patch(color='#27ae60', label='Collected - Documents (left semi-circle)'),
    mpatches.Patch(color='#2ecc71', label='Collected - Comments (right semi-circle)'),
    mpatches.Patch(color=color_not_collected, label='Not Collected', alpha=0.5),
]

# Add a few topic cluster colors to legend
selected_topics = ['Health & Medical', 'Food Safety & Agriculture', 'Transportation Safety', 
                   'Environmental & Natural Resources', 'Law Enforcement & Justice']
for topic in selected_topics:
    if topic in topic_cluster_colors:
        legend_elements.append(
            mpatches.Patch(edgecolor=topic_cluster_colors[topic], facecolor='white', 
                          linewidth=2, label=f'Topic: {topic}')
        )

fig.legend(handles=legend_elements, loc='lower center', ncol=4, 
          bbox_to_anchor=(0.5, -0.02), fontsize=10, frameon=True)

plt.tight_layout()
plt.savefig('agency_hierarchy.png', dpi=300, bbox_inches='tight')
print("Visualization saved to agency_hierarchy.png")

# Print summary statistics
print("\n=== SUMMARY STATISTICS ===")
collected_agencies = sum(1 for agency, data in agency_mapping.items() if data["collected"])
total_agencies = len(agency_mapping)
print(f"Total Agencies: {total_agencies}")
print(f"Collected: {collected_agencies}")
print(f"Not Collected: {total_agencies - collected_agencies}")

# Calculate actual totals
total_docs = sum(get_agency_data(a)[0] for a in agency_mapping.keys())
total_comments = sum(get_agency_data(a)[1] for a in agency_mapping.keys())
print(f"\nTotal Gov Documents: {total_docs:,}")
print(f"Total Comments: {total_comments:,}")

print("\n=== BY PARENT AGENCY ===")
for parent in sorted_parents:
    stats = parent_stats[parent]
    print(f"{parent:12s}: {stats['collected_count']:2d}/{stats['total_count']:2d} collected | "
          f"{stats['total_docs']:5,} docs | {stats['total_comments']:6,} comments")

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Wedge
from collections import defaultdict
import math

# Real data counts
gov_doc_counts = {'SEC': 16201, 'EPA': 15321, 'FAA': 10495, 'FDA': 8209, 'FEMA': 3605, 'FCC': 3508, 'CMS': 2273, 'DOE': 1555, 'NSF': 1432, 'FWS': 1415, 'USDA': 1339, 'USCBP': 1096, 'FS': 1081, 'CDC': 993, 'FDIC': 969, 'NHTSA': 924, 'OSHA': 910, 'AMS': 883, 'USCIS': 777, 'APHIS': 741, 'DHS': 538, 'MSHA': 371, 'FSIS': 268, 'TSA': 258, 'ICEB': 141, 'DOJ': 103, 'NIST': 33, 'NTIA': 23, 'NIH': 9}
aggregated_comment_counts = {'AMS': 4737, 'APHIS': 81502, 'CDC': 118967, 'CMS': 155055, 'DHS': 27, 'DOE': 477, 'DOJ': 392, 'EPA': 52681, 'FAA': 69318, 'FDA': 203401, 'USCIS': 63623, 'FEMA': 443, 'FS': 96652, 'FSIS': 6784, 'FWS': 82538, 'ICEB': 48, 'MSHA': 157, 'NHTSA': 70467, 'NIH': 2, 'NIST': 225, 'NTIA': 2604, 'OSHA': 12821, 'TSA': 97, 'USCBP': 102, 'USDA': 192}

agency_mapping = {
    "nih": {"head_agency": "HHS", "collected": True}, "fda": {"head_agency": "HHS", "collected": True},
    "cdc": {"head_agency": "HHS", "collected": True}, "cms": {"head_agency": "HHS", "collected": True},
    "acf": {"head_agency": "HHS", "collected": False}, "hrsa": {"head_agency": "HHS", "collected": False},
    "samhsa": {"head_agency": "HHS", "collected": False}, "acl": {"head_agency": "HHS", "collected": False},
    "usda": {"head_agency": "USDA", "collected": True}, "aphis": {"head_agency": "USDA", "collected": True},
    "ams": {"head_agency": "USDA", "collected": True}, "fsis": {"head_agency": "USDA", "collected": True},
    "fs": {"head_agency": "USDA", "collected": True}, "nrcs": {"head_agency": "USDA", "collected": False},
    "rma": {"head_agency": "USDA", "collected": False}, "fsa": {"head_agency": "USDA", "collected": False},
    "rd": {"head_agency": "USDA", "collected": False},
    "dhs": {"head_agency": "DHS", "collected": True}, "fema": {"head_agency": "DHS", "collected": True},
    "tsa": {"head_agency": "DHS", "collected": True}, "uscbp": {"head_agency": "DHS", "collected": True},
    "uscis": {"head_agency": "DHS", "collected": True}, "iceb": {"head_agency": "DHS", "collected": True},
    "cisa": {"head_agency": "DHS", "collected": False}, "uscg": {"head_agency": "DHS", "collected": False},
    "usss": {"head_agency": "DHS", "collected": False},
    "faa": {"head_agency": "DOT", "collected": True}, "nhtsa": {"head_agency": "DOT", "collected": True},
    "fmcsa": {"head_agency": "DOT", "collected": False}, "fra": {"head_agency": "DOT", "collected": False},
    "fta": {"head_agency": "DOT", "collected": False}, "phmsa": {"head_agency": "DOT", "collected": False},
    "marad": {"head_agency": "DOT", "collected": False},
    "osha": {"head_agency": "DOL", "collected": True}, "msha": {"head_agency": "DOL", "collected": True},
    "whd": {"head_agency": "DOL", "collected": False}, "ebsa": {"head_agency": "DOL", "collected": False},
    "ofccp": {"head_agency": "DOL", "collected": False},
    "nist": {"head_agency": "DOC", "collected": True}, "ntia": {"head_agency": "DOC", "collected": True},
    "noaa": {"head_agency": "DOC", "collected": False}, "census": {"head_agency": "DOC", "collected": False},
    "bis": {"head_agency": "DOC", "collected": False}, "uspto": {"head_agency": "DOC", "collected": False},
    "eda": {"head_agency": "DOC", "collected": False},
    "doj": {"head_agency": "DOJ", "collected": True}, "fbi": {"head_agency": "DOJ", "collected": False},
    "dea": {"head_agency": "DOJ", "collected": False}, "atf": {"head_agency": "DOJ", "collected": False},
    "bop": {"head_agency": "DOJ", "collected": False}, "usms": {"head_agency": "DOJ", "collected": False},
    "doe": {"head_agency": "DOE", "collected": True}, "nnsa": {"head_agency": "DOE", "collected": False},
    "fws": {"head_agency": "DOI", "collected": True}, "nps": {"head_agency": "DOI", "collected": False},
    "blm": {"head_agency": "DOI", "collected": False}, "bia": {"head_agency": "DOI", "collected": False},
    "usgs": {"head_agency": "DOI", "collected": False}, "boem": {"head_agency": "DOI", "collected": False},
    "epa": {"head_agency": "EPA", "collected": True}, "nsf": {"head_agency": "NSF", "collected": True},
    "sec": {"head_agency": "SEC", "collected": True}, "fdic": {"head_agency": "FDIC", "collected": True},
    "fcc": {"head_agency": "FCC", "collected": True},
}

def get_counts(code):
    key = code.upper()
    return gov_doc_counts.get(key, 0), aggregated_comment_counts.get(key, 0)

def log_size(value, lo, hi):
    if value <= 0: return lo
    return lo + (math.log10(value + 1) / math.log10(300000)) * (hi - lo)

# Group and stats
parent_to_children = defaultdict(list)
for ag, d in agency_mapping.items():
    parent_to_children[d["head_agency"]].append(ag)

parent_stats = {}
for p, ch in parent_to_children.items():
    td = sum(get_counts(c)[0] for c in ch)
    tc = sum(get_counts(c)[1] for c in ch)
    cc = sum(1 for c in ch if agency_mapping[c]["collected"])
    parent_stats[p] = {'total_docs': td, 'total_comments': tc, 'total_size': td+tc,
                        'collected_count': cc, 'total_count': len(ch)}

sorted_parents = sorted(parent_stats.keys(), key=lambda x: parent_stats[x]['total_size'], reverse=True)

# ============ LAYOUT IN ABSOLUTE UNITS (inches) ============
# We'll compute everything in inches then normalize to [0,1] at the end.

cols = 3
n_parents = len(sorted_parents)
n_rows = math.ceil(n_parents / cols)

# Size constants (inches)
PR_MIN_IN, PR_MAX_IN = 0.45, 1.1     # parent radius
CR_MIN_IN, CR_MAX_IN = 0.25, 0.65    # child collected radius
CR_UNCOL_IN = 0.25                     # uncollected child radius

P2C_GAP_IN = 2.2        # parent center to first child row center
C_ROW_STEP_IN = 1.8     # between child row centers
C_COL_SPREAD_IN = 5.5   # horizontal spread of children
C_LABEL_GAP_IN = 0.3    # node bottom to count label
C_LABEL_H_IN = 0.35     # count label height

STATS_GAP_IN = 0.6      # bottom of children to stats label
STATS_H_IN = 0.45       # stats label height
ROW_GAP_IN = 1.8        # gap between rows of trees

FIG_W = 32  # figure width in inches
X_MARGIN_IN = 1.5
COL_W = (FIG_W - 2 * X_MARGIN_IN) / cols

def child_radius_in(child):
    if agency_mapping[child]['collected']:
        d, c = get_counts(child)
        return log_size(d+c, CR_MIN_IN, CR_MAX_IN)
    return CR_UNCOL_IN

def parent_radius_in(parent):
    return log_size(parent_stats[parent]['total_size'], PR_MIN_IN, PR_MAX_IN)

def measure_tree_in(parent):
    """Returns (above, below) in inches from parent center."""
    pr = parent_radius_in(parent)
    above = pr

    children = parent_to_children[parent]
    n = len(children)
    if n == 0:
        below = pr + STATS_GAP_IN + STATS_H_IN
        return above, below

    ccols = min(3, n)
    crows = math.ceil(n / ccols)
    last_row_cy = P2C_GAP_IN + (crows - 1) * C_ROW_STEP_IN

    max_ext = 0
    for i, ch in enumerate(children):
        if i // ccols == crows - 1:
            cr = child_radius_in(ch)
            ext = cr + C_LABEL_GAP_IN + C_LABEL_H_IN
            max_ext = max(max_ext, ext)

    below = last_row_cy + max_ext + STATS_GAP_IN + STATS_H_IN
    return above, below

# Measure rows
row_above_in = []
row_below_in = []
for ri in range(n_rows):
    ma, mb = 0, 0
    for ci in range(cols):
        pi = ri * cols + ci
        if pi >= n_parents: break
        a, b = measure_tree_in(sorted_parents[pi])
        ma, mb = max(ma, a), max(mb, b)
    row_above_in.append(ma)
    row_below_in.append(mb)

total_h_in = sum(a + b for a, b in zip(row_above_in, row_below_in)) + (n_rows - 1) * ROW_GAP_IN
top_margin_in = 1.5
bot_margin_in = 0.5
fig_h = total_h_in + top_margin_in + bot_margin_in

# Compute y positions in inches (from top)
row_y_in = []  # y of parent center from bottom of figure
y_cursor = fig_h - top_margin_in - row_above_in[0]
for i in range(n_rows):
    row_y_in.append(y_cursor)
    if i < n_rows - 1:
        y_cursor -= (row_below_in[i] + ROW_GAP_IN + row_above_in[i + 1])

# Now convert everything to normalized [0,1] coordinates
def x_norm(inches_from_left):
    return inches_from_left / FIG_W

def y_norm(inches_from_bottom):
    return inches_from_bottom / fig_h

# Compute all positions in normalized coords
ppos = {}
cpos = {}
tree_bot_norm = {}

for idx, parent in enumerate(sorted_parents):
    row, col = idx // cols, idx % cols
    px_in = X_MARGIN_IN + (col + 0.5) * COL_W
    py_in = row_y_in[row]
    px, py = x_norm(px_in), y_norm(py_in)
    ppos[parent] = (px, py)
    cpos[parent] = []

    children = parent_to_children[parent]
    n = len(children)

    if n == 0:
        tree_bot_norm[parent] = y_norm(py_in - parent_radius_in(parent))
        continue

    ccols = min(3, n)
    cxstep = C_COL_SPREAD_IN / (ccols + 1)

    lowest_in = py_in
    for i, ch in enumerate(children):
        cr_i = i // ccols
        cc_i = i % ccols
        cx_in = px_in - C_COL_SPREAD_IN / 2 + (cc_i + 1) * cxstep
        cy_in = py_in - P2C_GAP_IN - cr_i * C_ROW_STEP_IN
        cpos[parent].append((x_norm(cx_in), y_norm(cy_in)))
        cr = child_radius_in(ch)
        node_bot_in = cy_in - cr - C_LABEL_GAP_IN - C_LABEL_H_IN
        lowest_in = min(lowest_in, node_bot_in)

    tree_bot_norm[parent] = y_norm(lowest_in)

# Convert radii to normalized (using fig dimensions)
def pr_norm(parent):
    r_in = parent_radius_in(parent)
    # Use average of x and y scale for circles
    return r_in / fig_h  # since aspect='equal' and ylim=1, radius in y-norm

def cr_norm(child):
    r_in = child_radius_in(child)
    return r_in / fig_h

cr_uncol_norm = CR_UNCOL_IN / fig_h

# ============ DRAWING ============
fig, ax = plt.subplots(1, 1, figsize=(FIG_W, fig_h))
fig.suptitle('Federal Agency Data Collection Hierarchy', fontsize=44, fontweight='bold', y=1 - 0.4/fig_h)

COL_UNCOL_FILL = '#95a5a6'

# Font sizes
F_PARENT = 24
F_CHILD = 14
F_CHILD_SMALL = 11
F_CHILD_UNCOL = 10
F_COUNTS = 11
F_STATS = 14
F_SUMMARY = 17
F_LEGEND = 16

# Draw connections
for parent in sorted_parents:
    px, py = ppos[parent]
    for i, ch in enumerate(parent_to_children[parent]):
        cx, cy = cpos[parent][i]
        ax.plot([px, cx], [py, cy], '-', color='#bdc3c7', alpha=0.5, linewidth=1.0, zorder=1)

# Draw parent nodes
for parent in sorted_parents:
    px, py = ppos[parent]
    total = parent_stats[parent]['total_size']
    r = pr_norm(parent)
    if total > 0:
        ax.add_patch(Wedge((px, py), r, 90, 270, facecolor='#2471a3', alpha=0.92, zorder=3, edgecolor='white', linewidth=2.5))
        ax.add_patch(Wedge((px, py), r, 270, 90, facecolor='#5dade2', alpha=0.92, zorder=3, edgecolor='white', linewidth=2.5))
    else:
        ax.add_patch(plt.Circle((px, py), r, facecolor='#aab7b8', alpha=0.7, zorder=3, edgecolor='white', linewidth=2.5))
    ax.text(px, py, parent, ha='center', va='center', fontsize=F_PARENT, fontweight='bold', color='white', zorder=4)

# Draw child nodes
for parent in sorted_parents:
    for i, ch in enumerate(parent_to_children[parent]):
        cx, cy = cpos[parent][i]
        data = agency_mapping[ch]
        if data['collected']:
            docs, comments = get_counts(ch)
            r = cr_norm(ch)
            ax.add_patch(Wedge((cx, cy), r, 90, 270, facecolor='#1e8449', alpha=0.9, edgecolor='white', linewidth=1.5, zorder=2))
            ax.add_patch(Wedge((cx, cy), r, 270, 90, facecolor='#58d68d', alpha=0.9, edgecolor='white', linewidth=1.5, zorder=2))
            fs = F_CHILD if len(ch) <= 4 else F_CHILD_SMALL
            ax.text(cx, cy, ch.upper(), ha='center', va='center', fontsize=fs, fontweight='bold', color='white', zorder=4)
            label_y = cy - r - C_LABEL_GAP_IN / fig_h
            ax.text(cx, label_y, f"{docs:,}d / {comments:,}c",
                    ha='center', va='top', fontsize=F_COUNTS, color='#333', zorder=4)
        else:
            r = cr_uncol_norm
            ax.add_patch(plt.Circle((cx, cy), r, facecolor=COL_UNCOL_FILL, alpha=0.5, edgecolor='white', linewidth=1, zorder=2))
            ax.text(cx, cy, ch.upper(), ha='center', va='center',
                    fontsize=F_CHILD_UNCOL, color='white', fontweight='bold', zorder=4)

# Stats label BELOW each tree
for parent in sorted_parents:
    px, _ = ppos[parent]
    st = parent_stats[parent]
    bot = tree_bot_norm[parent]
    line = f"{st['collected_count']}/{st['total_count']} collected"
    if st['total_size'] > 0:
        line += f"   ·   {st['total_docs']:,} docs   ·   {st['total_comments']:,} comments"
    ax.text(px, bot - STATS_GAP_IN / fig_h, line, ha='center', va='top', fontsize=F_STATS, zorder=4,
            bbox=dict(boxstyle='round,pad=0.4', facecolor='#f0f3f4', alpha=0.9, edgecolor='#bdc3c7', linewidth=0.8))

# Axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')
ax.set_aspect('equal')

# Summary box
td = sum(gov_doc_counts.values())
tc = sum(aggregated_comment_counts.values())
ta = len(agency_mapping)
co = sum(1 for d in agency_mapping.values() if d['collected'])
summary = f"Total Agencies: {ta}\nCollected: {co}  |  Not Collected: {ta - co}\n\nTotal Documents: {td:,}\nTotal Comments: {tc:,}"
ax.text(0.015, 0.998, summary, transform=ax.transAxes, fontsize=F_SUMMARY, va='top', fontfamily='monospace',
        bbox=dict(boxstyle='round', facecolor='#fef9e7', alpha=0.9, edgecolor='#d4ac0d', linewidth=1))

# Legend
legend_els = [
    mpatches.Patch(color='#2471a3', label='Parent: Documents (left half)'),
    mpatches.Patch(color='#5dade2', label='Parent: Comments (right half)'),
    mpatches.Patch(color='#1e8449', label='Collected: Documents (left half)'),
    mpatches.Patch(color='#58d68d', label='Collected: Comments (right half)'),
    mpatches.Patch(color=COL_UNCOL_FILL, alpha=0.5, label='Not Collected'),
]
ax.legend(handles=legend_els, loc='upper right', fontsize=F_LEGEND, frameon=True, fancybox=True, shadow=True, borderpad=1)

plt.subplots_adjust(top=0.99, bottom=0.003, left=0.01, right=0.99)
plt.savefig('agency_hierarchy_compact.png', dpi=150, bbox_inches='tight', facecolor='white')
print("Saved!")
```
